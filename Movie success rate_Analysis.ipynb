```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
```

```python
imdbdata=pd.read_csv('movie_success_rate.csv')
```

```python
imdbdata.shape
```

```python
imdbdata.head()
```

```python
imdbdata=imdbdata.rename(columns = {'Revenue (Millions)':'Revenue_Millions'})
imdbdata=imdbdata.rename(columns = {'Runtime (Minutes)':'Runtime_Minutes'})
```

```python
imdbdata.info()
```

```python
#check for null values
pd.isnull(imdbdata).sum()
```

```python
# drop null values
imdbdata.dropna(inplace=True)
```

```python
#DIRECTOR ANALYSIS :
imdbdata.Director.value_counts()[:10].plot.pie(autopct='%1.1f%%',figsize=(10,10))
plt.title('TOP 10 DIRECTORS OF MOVIES')
plt.show()
```

```python
#ACTOR ANALYSIS
imdbdata.Actors.value_counts()[:10].plot.pie(autopct='%1.1f%%',figsize=(10,10))
plt.title('TOP 10 ACTORS OF MOVIES')
plt.show()
```

```python
#YEAR ANALYSIS :
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="Year", y="Rating", data=imdbdata);
plt.title(' RATING BASED ON YEAR')
plt.show
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="Year", y="Votes", data=imdbdata);
plt.title(' VOTES BASED ON YEAR')
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="Year", y="Metascore", data=imdbdata);
plt.title(' METASCORE BASED ON YEAR')
```
